```python
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import StandardScaler
```

```python
model = tf.keras.models.load_model("cnn_lstm_model.h5")
```

```python
feature_ranges = {
    "HR": ( 100,200),
    "PULSE": ( 100,200),
    "RESP_x": ( 20,30),
    "SpO2": (100,120),
    "PLETH": (-1, 1),
    "II": (-1, 1)
}
```

```python
scaler = StandardScaler()
```

```python
def generate_live_data(time_steps=10):
    live_data = []
    for _ in range(time_steps):
        sample = [np.random.uniform(low, high) for _, (low, high) in feature_ranges.items()]
        live_data.append(sample)
    
    return np.array(live_data)
```

```python
live_input = generate_live_data(time_steps=10)
print(live_input)
```

```python
live_input = scaler.fit_transform(live_input) 
live_input = live_input.reshape((1, 10, len(feature_ranges)))  
print(live_input)
```

```python
prediction = model.predict(live_input)
```

```python
disease_status = "ðŸš¨ Possible Respiratory Disease" if prediction[0][0] > 0.5 else "âœ… No Disease"
```
